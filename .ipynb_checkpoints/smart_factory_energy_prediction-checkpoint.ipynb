```python
demo = df[['timestamp', 'zone1_temperature']].set_index('timestamp').copy()

# Create three series:
orig = demo['zone1_temperature']
ffill = orig.ffill()
linear = orig.interpolate(method='linear')
time = orig.interpolate(method='time')

plt.figure(figsize=(12, 6))
plt.plot(orig.index, orig, 'o', label='Original (with gaps)', markersize=3)
plt.plot(ffill.index, ffill, '-', label='Forward Fill')
plt.plot(linear.index, linear, '--', label='Linear Interpolation')
plt.plot(time.index, time, ':', label='Time-based Interpolation')
plt.legend()
plt.title('Imputation Comparison for zone1_temperature')
plt.xlabel('Timestamp')
plt.ylabel('Temperature (Â°C)')
plt.tight_layout()

plt.savefig('my_plot.png')
plt.show()


```

# Key Observations:
Negative Values Present:

1) The distribution shows readings below 0 (down to -500)
2) Energy consumption should theoretically never be negative
3) Likely indicates sensor errors or data recording issues

Bimodal Distribution:
1) Peaks around 0 and another around 500-1000

Extreme Positive Outliers:
1) Values extend up to 12,000 with very low frequency

```python
df.loc[df['equipment_energy_consumption'] < 0, 'equipment_energy_consumption'] = np.nan

# Then sort by timestamp and interpolate
df = df.sort_values('timestamp')
df['equipment_energy_consumption'] = df['equipment_energy_consumption'].interpolate(method='time')
```

```python
Q1 = df['equipment_energy_consumption'].quantile(0.25)
Q3 = df['equipment_energy_consumption'].quantile(0.75)
IQR = Q3 - Q1
upper_bound = Q3 + 1.5*IQR  # 3Ïƒ-equivalent for non-normal data

# Create clean series while preserving outlier info
df['energy_clean'] = np.where(df['equipment_energy_consumption'] > upper_bound,
                            upper_bound,
                            df['equipment_energy_consumption'])
```

```python
sns.histplot(df['energy_clean'], kde=True, bins=30)
plt.title('Distribution of Equipment Energy Consumption')
plt.savefig('my_plot.png')
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(10,6))
sns.boxplot(x=df['energy_clean'], flierprops=dict(marker='o', markersize=8))
plt.title("Energy Consumption Distribution with Outliers Marked")
plt.xlabel("Energy Consumption (kWh)")
plt.savefig('my_plot1.png')
plt.show()
```

```python
# Reset index (moves timestamps to a column)
df_plot = df.reset_index()

numeric_cols = df.select_dtypes(include=['float64']).columns
# Now create boxplots
for col in numeric_cols:
    Q1 = df_plot[col].quantile(0.25)
    Q3 = df_plot[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df_plot = df_plot[(df_plot[col] >= lower_bound) & (df_plot[col] <= upper_bound)]
plt.figure(figsize=(12, 6))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(len(numeric_cols)//3 + 1, 3, i)
    sns.boxplot(data=df_plot, y=col)  # Use data= parameter
    plt.title(f"Boxplot of {col}")
plt.tight_layout()
plt.show()



```
